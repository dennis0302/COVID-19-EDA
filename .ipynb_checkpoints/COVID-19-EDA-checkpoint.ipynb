# COVID-19 Exploratory Data Analysis

### COVID-19 has impacted all of us in 2020 and brought an unprecendented pandemic. In this notebook, I provide in-depth COVID-19 analysis in Canada.

### Please note that the dataset is from Statistics Canada and includes data from week 8 to week 47 in 2020.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline  

import seaborn as sns
```

```python
df = pd.read_csv('COVID19-eng.csv')
```

```python
df.head()
```

```python
df.info()
```

### We can see that the data has mostly categorical features other than features related to weeks. Thus, before analyzing and visualizing data, let's change numerical values into character values based on metadata provided from Statistics Canada. This will make it easier to understand plots.

```python
df.loc[df['Region'] == 1, 'Region'] = 'Atlantic'
df.loc[df['Region'] == 2, 'Region'] = 'Quebec'
df.loc[df['Region'] == 3, 'Region'] = 'Ontario/Nunavut'
df.loc[df['Region'] == 4, 'Region'] = 'Prairies'
df.loc[df['Region'] == 5, 'Region'] = 'BC/Yukon'
```

```python
df.loc[df['Age group'] == 1, 'Age group'] = '0 to 19 years'
df.loc[df['Age group'] == 2, 'Age group'] = '20 to 29 years'
df.loc[df['Age group'] == 3, 'Age group'] = '30 to 39 years'
df.loc[df['Age group'] == 4, 'Age group'] = '40 to 49 years'
df.loc[df['Age group'] == 5, 'Age group'] = '50 to 59 years'
df.loc[df['Age group'] == 6, 'Age group'] = '60 to 69 years'
df.loc[df['Age group'] == 7, 'Age group'] = '70 to 79 years'
df.loc[df['Age group'] == 8, 'Age group'] = '80 years or older'
df.loc[df['Age group'] == 99, 'Age group'] = 'Not Stated'
```

```python
df.loc[df['Transmission'] == 1, 'Transmission'] = 'Domestic acquisition'
df.loc[df['Transmission'] == 2, 'Transmission'] = 'International Travel'
df.loc[df['Transmission'] == 9, 'Transmission'] = 'Not Stated'
```

```python
df.loc[df['Gender'] == 1, 'Gender'] = 'Male'
df.loc[df['Gender'] == 2, 'Gender'] = 'Female'
df.loc[df['Gender'] == 9, 'Gender'] = 'Not stated'
```

```python
df.loc[df['Asymptomatic'] == 1, 'Asymptomatic'] = 'Yes'
df.loc[df['Asymptomatic'] == 2, 'Asymptomatic'] = 'No'
df.loc[df['Asymptomatic'] == 9, 'Asymptomatic'] = 'Not Stated'
```

```python
df.loc[df['Death'] == 1, 'Death'] = 'Died'
df.loc[df['Death'] == 2, 'Death'] = 'Not Died'
df.loc[df['Death'] == 9, 'Death'] = 'Not Stated'
```

```python
df.loc[df['Hospital status'] == 1, 'Hospital status'] = 'Hospitalized/Intensive Care'
df.loc[df['Hospital status'] == 2, 'Hospital status'] = 'Hospitalized/Not Intensive Care'
df.loc[df['Hospital status'] == 3, 'Hospital status'] = 'Not Hospitalized'
df.loc[df['Hospital status'] == 9, 'Hospital status'] = 'Not Stated'
```

```python
df.loc[df['Recovered'] == 1, 'Recovered'] = 'Recovered'
df.loc[df['Recovered'] == 2, 'Recovered'] = 'Not Recovered'
df.loc[df['Recovered'] == 9, 'Recovered'] = 'Not Stated'
```

### First, let's plot the weekly confirmed cases in 2020.

```python
df_by_week = df.groupby('Episode week').agg({'Case identifier number':'count'}).reset_index()
```

```python
# exclude values with 99 (no date is available)
df_by_week = df_by_week[:-1]
```

```python
fig = plt.figure()
fig.set_size_inches(15, 10)

plt.title('Number of COVID-19 Cases',size=15)
sns.barplot(x=df_by_week['Episode week'],y=df_by_week['Case identifier number'])
plt.xlabel('Week (up to mid November)', fontsize=18)
plt.ylabel('Number of weekly cases', fontsize=16)
plt.show()
```

### We can see that cases surged in week 14-15 slowed down until week 33. Then, we can notice that cases significantly rise after week 35.

### Now, divide the data by regions to see which provinces/territories contribute the most to rising confirmed cases.

```python
df_by_region = df.groupby('Region').agg({'Case identifier number':'count'}).reset_index()
df_by_region
```

```python
sns.set()
fig = plt.figure()
fig.set_size_inches(14, 10)

sns.barplot(x=df_by_region['Region'], y=df_by_region['Case identifier number'], palette='hls')
plt.title('Number of COVID-19 Cases by Region', size=15)
plt.xlabel('Region', fontsize=18)
plt.ylabel('Number of cases', fontsize=16)
plt.show()
```

### As expected, two most populous provinces in Canada, Ontario and Quebec, had the most number of confirmed cases. Although Statistics Canada grouped Ontario and Nunavut together, it is safe to assume that confirmed cases from Nunavut is negligible due to the small population.

### To find out why COVID-19 cases remained consistent in Canada, we would like to know how COVID-19 was able to transmit and spread in Canada.

```python
df_by_transmission = df.groupby('Transmission').agg({'Case identifier number':'count'}).reset_index()
df_by_transmission
```

```python
sns.set()
fig = plt.figure()
fig.set_size_inches(14, 10)

plt.title('COVID-19 Transmission Types in Canada', size=15)
sns.countplot(x=df['Transmission'], hue=df['Region'], palette='Paired')
plt.xlabel('Transmission type', fontsize=18)
plt.ylabel('Number of cases', fontsize=16)
plt.show()
```

### From looking at this bar chart, we can see that most of confirmed cases are transmitted within Canada. Transmissions from international travel are significantly lower. However, as COVID-19 is extremely contagious, it is probable that small number of confirmed cases from international travel in early pandemic days have accelerated the spread of COVID-19 in Canada.

### Now, let's understand more about COVID-19 by analyzing demographics!

```python
df_by_gender = df.groupby('Gender').agg({'Case identifier number':'count'}).reset_index()
df_by_gender
```

```python
c = ['pink','sandybrown','gold']
plt.figure(figsize=(20,15))
plt.title('Number of COVID-19 Cases by Gender', size=15)
plt.pie(df_by_gender['Case identifier number'], colors=c,shadow=True)
plt.legend(df_by_gender['Gender'], loc='best', fontsize=12)
plt.show()
```

### From the pie chart, we can see that females had slightly more confirmed cases than males.

### Would age matter for getting COVID-19?

```python
df_by_age = df.groupby('Age group').agg({'Case identifier number':'count'}).reset_index()
df_by_age
```

```python
fig = plt.figure()
fig.set_size_inches(13, 10)

plt.title('Number of COVID-19 Cases by Age Group',size=15)
b = plt.bar(df_by_age['Age group'],df_by_age['Case identifier number'])
b[1].set_color('red')
plt.xlabel('Age group', fontsize=18)
plt.ylabel('Number of cases', fontsize=16)
plt.show()
```

### The bar chart tells us that younger age groups are more vulnerable to COVID-19. At this point, you might be also wondering if COVID-19 is more severe for younger or older age groups.

```python
df_asymptomatic = df.groupby('Asymptomatic').agg({'Case identifier number':'count'}).reset_index()
df_asymptomatic
```

```python
sns.set()

fig = plt.figure()
fig.set_size_inches(13, 10)
sns.catplot(y="Age group", hue="Asymptomatic", kind="count", palette="rocket", alpha=.6, height=10, data=df)
plt.xlabel('Number of cases', fontsize=18)
plt.ylabel('Age group', fontsize=16)
plt.title('Number of COVID-19 Cases by Age Group and Sympton Signs',size=15)
plt.show()
```

### Aside from patients who didn't state their symptoms, there are more patients who showed signs of symtoms than patients without symptoms. Although younger age groups are more vulnerable to COVID-19, they are more likely to have no symptons except for 0 to 19 years age group. Older age groups are more likely to have signs of symptons.

```python
df_status = df.groupby('Hospital status').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_status
```

```python
sns.set()

fig = plt.figure()
fig.set_size_inches(13, 10)
sns.catplot(y="Age group", hue="Hospital status", kind="count", palette="dark", alpha=.6, height=10, data=df)
plt.xlabel('Number of patients', fontsize=18)
plt.ylabel('Age group', fontsize=16)
plt.title('Number of Patients Hospitalized by Age Group',size=15)
plt.show()
```

### Although most of patients are not hospitalized, older age groups are more likely to be hospitalized.

### Let's take a look at what symptoms patients have experienced. Statistics Canada outlined 12 symptoms in the dataset. It includes cough, fever, chills, sore throat, runny nose, shortness of breath, nausea, headache, weakness, pain, irritability and diarrhea.

```python
df_sym1 = df.groupby('Symptom, cough').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym2 = df.groupby('Symptom, fever').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym3 = df.groupby('Symptom, chills').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym4 = df.groupby('Symptom, sore throat').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym5 = df.groupby('Symptom, runny nose').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym6 = df.groupby('Symptom, shortness of breath').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym7 = df.groupby('Symptom, nausea').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym8 = df.groupby('Symptom, headache').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym9 = df.groupby('Symptom, weakness').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym10 = df.groupby('Symptom, pain').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym11 = df.groupby('Symptom, irritability').agg({'Case identifier number':'count'}).reset_index()[:-1]
df_sym12 = df.groupby('Symptom, diarrhea').agg({'Case identifier number':'count'}).reset_index()[:-1]
```

```python
plot = plt.figure()

cough = df_sym1['Case identifier number'].tolist()
fever = df_sym2['Case identifier number'].tolist()
chills = df_sym3['Case identifier number'].tolist()
sore_throat = df_sym4['Case identifier number'].tolist()
runny_nose = df_sym5['Case identifier number'].tolist()
shortness_of_breath = df_sym6['Case identifier number'].tolist()
nausea = df_sym7['Case identifier number'].tolist()
headache = df_sym8['Case identifier number'].tolist()
weakness = df_sym9['Case identifier number'].tolist()
pain = df_sym10['Case identifier number'].tolist()
irritability = df_sym11['Case identifier number'].tolist()
diarrhea = df_sym12['Case identifier number'].tolist()

index = ['Symptomatic','Asymptomatic']

df_sympton_signs = pd.DataFrame({'cough': cough, 'fever': fever,'chills':chills,'sore_throat':sore_throat,\
                   'runny_nose':runny_nose,'shortness_of_breath':shortness_of_breath,\
                  'nausea':nausea,'headache':headache,'weakness':weakness,'pain':pain,\
                  'irritability':irritability,'diarrhea':diarrhea}, index=index)
```

```python
df_sympton_signs
```

```python
fig = plt.figure()
fig.set_size_inches(14, 10)

ax = df_sympton_signs.plot.bar(rot=0,figsize=(14,10))
ax.legend(bbox_to_anchor=(1.0, 1.0))
plt.title('Number of Patients with/without 12 Symptoms',size=15)
plt.ylabel('Number of patients', fontsize=16)
plt.show()
```

### From the table and bar chart, we can see that most common symptoms patients have are cough, headache, sore throat, fever and chills.

### Let's take a look at how many patients were able to recover from COVID-19.

```python
df_recovered = df.groupby('Recovered').agg({'Case identifier number':'count'}).reset_index()
```

```python
df_recovered
```

```python
c = ['navajowhite', 'lightcoral','orange']
plt.figure(figsize=(20,15))
plt.title('Recovered Status from COVID-19', size=15)
plt.pie(df_recovered['Case identifier number'], colors=c,shadow=True)
plt.legend(df_recovered['Recovered'], loc='best', fontsize=12)
plt.show()
```

### The pie chart tells us that most of people were able to recover from COVID-19.

### At this point, It would be interesting to know how many people were able to recover from COVID-19 if they were hospitalized.

```python
sns.set()
fig = plt.figure()
fig.set_size_inches(14, 10)

sns.catplot(y="Hospital status", hue="Recovered", kind="count",
            palette="pastel", edgecolor=".6", height=10,
            data=df)
plt.title('Recovered Status by Hospital Status', size=15)
plt.xlabel('Number of patients', fontsize=18)
plt.ylabel('Hospital status', fontsize=16)
plt.show()
```

### Given the small samples of data from hospitalized patients, we can see that the hospitalized patients are not likely to be recovered.

```python
sns.set()
fig = plt.figure()
fig.set_size_inches(14, 10)

sns.catplot(y="Age group", hue="Death", kind="count",
            palette="Set2", edgecolor=".6", height=10,
            data=df)
plt.title('Death from COVID-19 by Age Group', size=15)
plt.xlabel('Number of patients', fontsize=18)
plt.ylabel('Age group', fontsize=16)
plt.show()
```

### As older age groups have weaker immune system, the bar chart shows that COVID-19 can be deadly for older age groups.
